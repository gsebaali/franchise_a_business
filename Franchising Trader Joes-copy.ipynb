Geopy allows for locating the coordinates of addresses, cities, countries, and landmarks across the globe using third-party geocoders and other data sources. RateLimiter is used to perform bulk operations to handle error responses and timeouts and adding delays when needed.

__Data Extraction__
 The locations are treated differently as follows:

- For locations with a Trader Joe's store available, I fetch the coordinates of a Trader Joe's within that city.
```python
# sample n for 1 indicator and n for 0 indicator
pd_cities_1= pd_cities[pd_cities['Indicator']==1]
#pd_cities_1 = pd_cities_1.sample(n=70)

pd_cities_0= pd_cities[pd_cities['Indicator']==0]
#pd_cities_0 = pd_cities_0.sample(n=70)
```

```python
# use geocoder to fetch the lat and lng of cities with locations that do not have a Trader Joe's
from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim

lat=[]
lng=[]


for city in pd_cities_0['Address']:
    address = city
    #print(address)
    
    geolocator = Nominatim(user_agent="ca_explorer",timeout=4)
    #RateLimiter allows to perform bulk operations while gracefully handling error responses and adding delays when needed.
    #to avoid: urlopen error timed
    geocode = RateLimiter(geolocator.geocode, min_delay_seconds=4)
    location = geolocator.geocode(address)
    lat.append(location.latitude)
    lng.append(location.longitude)
    
    
pd_cities_0.loc[:,'Latitude']=lat
pd_cities_0.loc[:,'Longitude']=lng
```

```python
#use geocoder to fetch the lat and lng of cities that have a Trader Joes
lat=[]
lng=[]


for city in pd_cities_1['Address']:
    address = 'Trader Joe\'s, ' + city
    #print(address)
    
    geolocator = Nominatim(user_agent="ca_explorer",timeout=4)
    #RateLimiter allows to perform bulk operations while gracefully handling error responses and adding delays when needed.
    #to avoid: urlopen error timed
    geocode = RateLimiter(geolocator.geocode, min_delay_seconds=4)
    location = geolocator.geocode(address)
    
    #some trader joe locations arent found using geocoder
    if (location == None):
        lat.append(0)
        lng.append(0)
    else:
        
        lat.append(location.latitude)
        lng.append(location.longitude)
    
    
pd_cities_1.loc[:,'Latitude']=lat
pd_cities_1.loc[:,'Longitude']=lng
```

#### Data Source 4 API Foursquare for feature set of venues

The Foursquare API allows application developers to interact with the Foursquare platform, a social location service that allows users to explore the world around them. The API itself is a RESTful set of addresses to which we send requests to explore a geographical location at a radius and get back JSON file. I construct a URL to send a request to the API to explore the geocoordinates built in the previous step. I set the radius to 1000m and LIMIT to 100 to explore a maximum of 100 different venues at a radius of 1000m from the location.
Below are the steps to collect the count of venue categories surrounding a geographical location.

__Data Extraction__
- Use my Foursquare developer account to authenticate and send requests to Foursquare API
- Construct a URL in the following format setting the radius to 1000m and limit to 100: > `https://api.foursquare.com/v2/venues/`**explore**`?client_id=`**CLIENT_ID**`&client_secret=`**CLIENT_SECRET**`&ll=`**LATITUDE**`,`**LONGITUDE**`&v=`**VERSION**`&limit=`**LIMIT** 
- Send GET request and examine results

__Data Wrangling__
- Get venue category from JSON
- Group venues by category and count
- Append resut to the dataframe where each location will have one row corresponding to counts of each venue category
- Replace NaN with 0 to indicate a 0 count for those venues that do not apply to certain locations

```python
# setup foursquare
CLIENT_ID = 'FGAZJB5V2EAYBCSGTJ33WU1J442EAWRUQSCITAVQJOAT30LO' #  Foursquare ID
CLIENT_SECRET = 'GFSPWKLHLLJ0U2HC3UETY4P3AULEZZKWBFM1YQZ543O0OHMM' #  Foursquare Secret
VERSION = '20180604'
```

```python
# create an empty dataframe for the venue's count
venues = pd.DataFrame()
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# send URL requests to Foursqaure API, obtain the JSON results, normalize data and append to venues dataframe
from pandas.io.json import json_normalize
radiusN=1000
LIMITN=100


for lat, lng, neigh in zip(pd_cities_n['Latitude'], pd_cities_n['Longitude'],  pd_cities_n['Name']):

    #print(neigh)
    url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET,  lat, lng, VERSION, radiusN, LIMITN)
    results = requests.get(url).json()
    
    if len(results['response'])==0:
        continue
    elif len(results['response']['groups'][0]['items'])==0:#nothing venues to explore
        continue
    else:
    
        items = results['response']['groups'][0]['items']
    
        dataframe = json_normalize(items) # flatten JSON
        if (dataframe.empty ):
            continue
        else:
            # filter columns
            filtered_columns = ['venue.name', 'venue.categories'] + [col for col in dataframe.columns if col.startswith('venue.location.')] + ['venue.id']
            dataframe_filtered = dataframe.loc[:, filtered_columns]

            # filter the category for each row
            dataframe_filtered['venue.categories'] = dataframe_filtered.apply(get_category_type, axis=1)
    

            # clean columns
            dataframe_filtered.columns = [col.split('.')[-1] for col in dataframe_filtered.columns]
            
            
            # get count per category
            a = dataframe_filtered['categories'].value_counts().to_frame().transpose()
            a['Name']=neigh
        

            venues=venues.append(a,ignore_index = True, sort=True)

```

## Methodology: Regression Analysis Modeling and Evaluation

In this section, I will discuss the methodologies used for training machine learning models. For splitting data into test and train, I use leave-one out cross validation method which provides train/test indices to split data in train/test sets. Each sample is used once as a test set (singleton) while the remaining samples form the training set.

__Modeling__ As the goal of this report is to predict the probability of a location to be that of a Trader Joe's, the models trained in this report are the following regression models:
- Logistic Regression
- SVM
- SVM with GridSearchCV

__Evaluation__ To evaluate each of the models, I use the following metrics [6]: 

- __Precision__ 
This is a measure of the accuracy provided that a class label has been predicted. It is defined by: precision = TP / (TP + FP) where TP is True Positives and FP is False Positives.

- __Recall__ 
This is true positive rate. It is defined as: Recall =  TP / (TP + FN) where TP is True Positives and FN is False Negatives.

- __Jaccard index for accuracy evaluation:__
This metric is defined as the size of the intersection divided by the size of the union of two label sets. If the entire set of predicted labels for a sample strictly match with the true set of labels, then the subset accuracy is 1.0; otherwise it is 0.0.

- __F1 score for accuracy evaluation:__
This metric is based on the precision and recall of a label. The F1 score is the harmonic average of the precision and recall. A perfect precision and recall correspond to an F1 score of 1. It is a good way to show that a classifer has a good value for both recall and precision. 

- __Log loss for evaluation:__
This measures the performance of a classifier where the predicted output is a probability value between 0 and 1.

### Logistic Regression

<a id="ref1"></a>

In order to estimate the class of a location, it is required to know what would be the most probable class for that location: a Trader Joe's location or not. For this, I use Logistic Regression which is a variation of Linear Regression, useful when the observed dependent variable, y, is categorical. It predicts the probability of the class label as a function of the independent variables. This method fits a special s-shaped curve by taking the linear regression and transforming the numeric estimate into a probability with the following function, which is called sigmoid function 𝜎:

$$
ℎ_\theta(𝑥) = \sigma({\theta^TX}) =  \frac {e^{(\theta_0 + \theta_1  x_1 + \theta_2  x_2 +...)}}{1 + e^{(\theta_0 + \theta_1  x_1 + \theta_2  x_2 +\cdots)}}
$$
Or:
$$
ProbabilityOfaClass_1 =  P(Y=1|X) = \sigma({\theta^TX}) = \frac{e^{\theta^TX}}{1+e^{\theta^TX}} 
$$

In this equation, ${\theta^TX}$ is the regression result (the sum of the variables weighted by the coefficients), `exp` is the exponential function and $\sigma(\theta^TX)$ is the sigmoid or [logistic function](http://en.wikipedia.org/wiki/Logistic_function), also called logistic curve. It is a common "S" shape (sigmoid curve).


The objective of Logistic Regression algorithm, is to find the best parameters θ, for $ℎ_\theta(𝑥)$ = $\sigma({\theta^TX})$, in such a way that the model best predicts the class of each case.

__Logistic Regression Tuning Parameters__ I use Logistic Regression method from Scikit-learn and tune C (C=1 found to be optimal), the parameter that indicates the inverse of regularization strength where regularization is a technique used to solve the overfitting problem in machine learning models. Smaller values specify stronger regularization. 

```python
loo = LeaveOneOut()
loo.get_n_splits(X)
np_ytest=[]
np_ytrain=[]
np_yhat_train=[]
np_yhat=[]
np_yhat_prob=[]
np_yhat_prob_complete=[]
np_yscore_lr=[]

for train_index, test_index in loo.split(X):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    

    # apply logistic regression
    LR = LogisticRegression(C=1, solver='liblinear', tol=1e-24).fit(X_train,y_train)
    y_score_lr = LR.decision_function(X_test)

    np_yscore_lr.append(y_score_lr[0])
    # predict using logistic regression
    yhat = LR.predict(X_test)
    yhat_train = LR.predict(X_train)
    
    # predict_proba returns estimates for all classes, ordered by the label of classes.
    yhat_prob = LR.predict_proba(X_test)
    
    # update output lists
  
    np_yhat_prob_complete.append(yhat_prob[0])
    np_yhat.append(yhat)
    np_ytest.append(y_test[0])
    #if y_test==1:
    #    np_yhat_prob.append(yhat_prob[0][1])
    #else:
    #    np_yhat_prob.append(yhat_prob[0][0])
    np_yhat_prob.append(yhat_prob[0][1])


```

### SVM

Support vector machines (SVMs), a set of supervised learning methods used for classification, regression and outliers detection, are also trained in this report.

The advantages of support vector machines are [5]:

- Effective in scenarios where number of dimensions is greater than the number of samples.
- Versatile in the sense that different Kernel functions are available for the decision function.

The disadvantages of support vector machines include [5]:

- SVMs do not directly provide probability estimates, these are calculated using an expensive five-fold cross-validation.


The SVM algorithm offers a choice of kernel functions for performing its processing. The mathematical function used for the transformation is known as the kernel function, and can be of different types, such as:

- Linear
- Polynomial
- Radial basis function (RBF)
- Sigmoid

The best performing model was trained using a sigmoid kernel.


```python
from sklearn import svm

looo = LeaveOneOut()
looo.get_n_splits(X)
np_ytest_svm=[]
np_yhat_svm=[]
np_yhat_prob_svm=[]
np_yhat_prob_complete_svm=[]
np_yscore_svm=[]

for train_index, test_index in looo.split(X):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
     
    # apply SVM
    clf_svm= svm.SVC(probability=True)
    clf_svm.fit(X_train, y_train) 
    yhat_svm = clf_svm.predict(X_test)
    
    y_score_svm = clf_svm.decision_function(X_test)
    np_yscore_svm.append(y_score_svm[0])
    
    
    yhat_prob = clf_svm.predict_proba(X_test)
    np_yhat_prob_complete_svm.append(yhat_prob[0])

    np_yhat_svm.append(yhat_svm)
    np_ytest_svm.append(y_test[0])
    

    np_yhat_prob_svm.append(yhat_prob[0][1])

    
```

### SVM and hyperparameter tuning

SVM is optimized by using sklearn.model_selection.GridSearchCV and cross-validation to tune the hyper-parameters:
- __kernel:__ specifies the kernel function to be used to map observations to a high dimension for classification. It can take one of the following forms:
    - ‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’ or a callable.
- __C:__ specifies the regularization parameter. The strength of the regularization is inversely proportional to C.
    - C>0 and I check for the following values of C [0.1, 0.5, 1, 5, 10]
- __gamma:__ specifies the kernel coefficient for ‘rbf’, ‘poly’ and ‘sigmoid’.
    - [0.0001, 0.001, 0.005, 0.1, 1, 3, 5]

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

loo = LeaveOneOut()
loo.get_n_splits(X)


 
for train_index, test_index in loo.split(X):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    parameters = {'kernel':('linear', 'rbf','sigmoid'), 'C':[0.1, 0.5, 1, 5, 10]
         ,  'gamma': [0.0001, 0.001, 0.005, 0.1, 1, 3, 5]
    }
    svc = svm.SVC()
    clf = GridSearchCV(svc, parameters)
    clf.fit(X_train, y_train)
```
